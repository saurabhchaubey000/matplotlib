```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Function based
x=np.arange(0,10)
```

```python
x
```

```python
y=x**2
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color='green',linestyle='dotted')
```

```python
plt.plot(x,y,color='green',linestyle='dashed')
```

```python
plt.plot(x,y,color='green',linestyle='-.')
```

```python
plt.plot(x,y,color='green',linestyle='-.',linewidth=5)
```

```python
plt.plot(x,y,color='green',linestyle='-.',linewidth=5,alpha=0.75)
```

```python
plt.plot(x,y,color='green',linestyle='-.',linewidth=5,alpha=0.75)
plt.xlabel('X value')
plt.ylabel('Y value')
plt.title('Chart Title')
```

```python
plt.plot(x,y,color='green',linestyle='-',linewidth=5,alpha=0.75,marker='+',markersize=15)
plt.xlabel('X value')
plt.ylabel('Y value')
plt.title('Chart Title')
```

```python
plt.plot(x,y,color='green',linestyle='-',linewidth=5,alpha=0.50,marker='o',markersize=15,markerfacecolor='red')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.title('Chart Title')
```

```python
z=x+10
print(z)
plt.plot(x,y,label='X',color='green')
plt.plot(y,x,label='Y',color='purple')
plt.plot(x,z,label='Z',color='red')
plt.legend(loc='best')
```

```python
# OO (Object Oriented method) 
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.9])
ax1=fig.add_axes([0.5,0.5,0.3,0.3])
ax.plot(x,y)
ax1.plot(y,x)
```

```python
fig1=plt.figure()
ax=fig1.add_axes([0.1,0.1,0.8,0.9])
ax1=fig1.add_axes([0.2,0.5,0.3,0.3])
ax.plot(x,y,'r--')
ax.set_xlabel('X value')
ax.set_ylabel('Y value')
ax.set_title('Outer Graph')
ax1.plot(y,x,'b-.')
ax1.set_ylabel('Y value')
ax1.set_xlabel('X value')
ax1.set_title('Inner graph')
```

```python
fig2=plt.figure(figsize=(10,4),dpi=50) #figsize=(width,height)
ax=fig2.add_axes([0,0,.8,.8])
ax.plot(x,y)
```
