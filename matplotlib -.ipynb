```python
#bar chart
import numpy as np
comp=['Google','Facebook','Amazon','Yahoo']
cmp= np.arange(len(comp))
rev=[7000,6000,12000,8500]
prof=[700,600,120,300]
cmp
```

```python
plt.xticks(cmp,comp)
plt.bar(cmp,rev,color='g')
plt.show()
```

```python
plt.xticks(cmp,comp)
plt.bar(cmp-.2,rev,color='g')
plt.bar(cmp+0.2,prof,color='r')
plt.show()
```

```python
plt.xticks(cmp,comp)
plt.bar(cmp-.2,rev,color='g',width=.4,label='Revenue')
plt.bar(cmp+0.2,prof,color='r',width=.4,label='Profit')
plt.legend()
plt.show()
```

```python
# Pie chart
ex_type=['Home Rent','Food','Phone/Internet bill','Car','Other Utilities']
ex_val=[1400,600,300,400,600]
```

```python
plt.pie(ex_val)
```

```python
plt.pie(ex_val,labels=ex_val)
plt.show()
```

```python
plt.pie(ex_val,labels=ex_val,autopct='%0.0f%%')
plt.show()
```

```python
plt.pie(ex_val,labels=ex_val,autopct='%0.5f%%')
plt.show()
```

```python
plt.pie(ex_val,labels=ex_val,autopct='%0.0f%%',radius=2)
plt.show()
```

```python
plt.pie(ex_val,labels=ex_val,autopct='%0.0f%%',explode=[0,.2,0,0,0.2])
plt.show()
```

```python
plt.pie(ex_val,labels=ex_val,autopct='%0.0f%%',startangle=180)
plt.show()
```
