```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig=plt.figure()
x1=[10,20,30,40,50,60]
y1=[1000,2000,1500,2100,2700,1700]
y2=[1200,1800,1600,2000,2500,1500]
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x1,y1)
ax.plot(x1,y2)
fig.savefig(r'e:\my_first.jpeg',dpi=200)
```

```python
fig,ax=plt.subplots()
ax.plot(x1,y1)
```

```python
fig1,ax1=plt.subplots(nrows=1,ncols=2)
ax1[0].plot(x1,y1)
ax1[1].plot(x1,y1)
plt.tight_layout()
```

```python
fig2,ax2=plt.subplots(nrows=1,ncols=2)
for x in ax2:
    x.plot(x1,y1)
```
